```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('C:\\Users\\sys8\\Downloads\\Iris.csv')
```

```python
df.head()
```

```python
#checking missing values

df[df.isnull().any(axis=1)]
```

```python
df['Species']
```

```python
lb_make = LabelEncoder()
df["Species1"] = lb_make.fit_transform(df["Species"])
df[["Species", "Species1"]].head(100)
```

```python
y = df['Species1']
```

```python
#splitting test train

x_train,x_test,y_train,y_test=train_test_split(df,y,test_size=0.33,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
model = KNeighborsClassifier(n_neighbors=6) # default value for n_neighbors is 5
```

```python
# Train the model using the training sets and check score
model.fit(x_train, y_train)
```

```python
#Predict Output
predicted= model.predict(x_test)
predicted
```

```python
#accuracy scores

model.score(x_test,y_test)
```

```python
pd.concat([x_test, y_test, pd.Series(predicted, name='Predicted', index=x_test.index)], 
          ignore_index=False, axis=1)
```
