```python

from flask import Flask, make_response, request, jsonify
# build the flask app
app = Flask(__name__)

def avg_price(br,bathR,sqftliving,yrbuilt,zipco):
    import pandas as pd
    import numpy as np
    import os
    from sklearn.metrics import mean_squared_error 
    from math import sqrt
    from sklearn.model_selection import train_test_split
    from sklearn import neighbors
    from sklearn import ensemble
    
    os.chdir("C:\\Users\\sys6\\Documents")
    data = pd.read_csv("house_prediction_bot.csv")
    data_new = data[['bedrooms','bathrooms','sqft_living','yr_built','zipcode','price']]
    y = data_new['price']
    data_new1 = data_new.drop('price',axis =1)
    x_train,x_test,y_train,y_test=train_test_split(data_new1,y,test_size=0.33,random_state=42)
    model = ensemble.RandomForestRegressor()
    model.fit(x_train,y_train)
    xyz = [br,bathR,sqftliving,yrbuilt,zipco]
    new_1 = np.array(xyz).reshape(1,-1)
    
    if 1900<yrbuilt<2019 and zipco in range(98001,98200):
        
        value =  model.predict(new_1)
        predict = 'Average cost is : $'+str(value[0])
        return predict[0:]
    else:
        predict = ['Please enter valid year or zipcode']
        return predict[0]

    
# definition of the results function
def results():
    req = request.get_json(force=True)
    action = req.get('queryResult').get('action')
    location_name = req.get('queryResult').get('parameters').get('any')
    zipcode_name = req.get('queryResult').get('parameters').get('zip-code')
    home_req = req.get('queryResult').get('parameters').get('home')
    br =  req.get('queryResult').get('parameters').get('bedroom')
    bathR = req.get('queryResult').get('parameters').get('bathroom')
    sqftliving = req.get('queryResult').get('parameters').get('sqrtlliving')
    yrbuilt = req.get('queryResult').get('parameters').get('number')
    name = req.get('queryResult').get('parameters').get('given-name')
    #print(type(br),type(bathR),type(sqftliving),type(yrbuilt),type(zipcode_name))
    selling_price = avg_price(br,bathR,sqftliving,yrbuilt,int(zipcode_name))
    
    if action == "house_sell.house_sell-yes":
        #selling_price = buyhouse_.avg_price(br,bathR,sqftliving,yrbuilt,zipcode_name)

        result = {} # an empty dictionary

        # fulfillment text is the default response that is returned to the dialogflow request
        result["fulfillmentText"] = selling_price #'Thank you '+str(name)+' for providing all the details. My estimated cost to sell your house is between: $' + str(selling_price-selling_price*0.01)+' to $'+str(selling_price)

        # jsonify the result dictionary
        # this will make the response mime type to application/json
        result = jsonify(result)

        # return the result json
        return make_response(result)
    elif action == 'house_buys.house_buys-yes':
        #selling_price = buyhouse_.avg_price(br,bathR,sqftliving,yrbuilt,zipcode_name)
        result = {} # an empty dictionary
#buyhouse234.avg_price(br,bathR,sqftliving,yrbuilt,zipcode_name)
        # fulfillment text is the default response that is returned to the dialogflow request
        result["fulfillmentText"] = selling_price #'Thank you '+str(name)+' for giving all the details. My estimated approximate value to buy a house for these requirements is between : $'+ str(selling_price-selling_price*0.01)+ ' to $'+str(selling_price)
        # jsonify the result dictionary
        # this will make the response mime type to application/json
        result = jsonify(result)
        # return the result json
        return make_response(result)
   
        
   
   
  #default route for the webhook
# it accepts both the GET and POST methods
@app.route('/webhook', methods=['GET', 'POST'])
def index():
    # calling the result function for response
    return results()

# call the main function to run the flask ap
if __name__ == '__main__':
   app.run(debug=False)
```
